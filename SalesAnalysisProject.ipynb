```python
# Data preparation
```

```python
# Ensure OrderDate is in datetime format
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

```python
# Create a new column ('order_month') with the full month name
```

```python
df['order_month'] = df['OrderDate'].dt.strftime('%B')
```

```python
# Create order_quarter column (1, 2, 3, 4)
```

```python
df['order_quarter'] = df['OrderDate'].dt.quarter
```

```python
# Create order_year column
```

```python
df['order_year'] = df['OrderDate'].dt.year
```

```python
# Create order_day_name column
```

```python
df['order_day_name'] = df['OrderDate'].dt.day_name()
```

```python
# Ensure ShipDate is in datetime format
```

```python
df['ShipDate'] = pd.to_datetime(df['ShipDate'])
```

```python
# Create ShippingTime column
```

```python
df['ShippingTime'] = (df['ShipDate'] - df['OrderDate']).dt.days
```

```python
# Change Date format in OrderDate & ShipDate columns
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate']).dt.strftime('%d-%m-%Y')
```

```python
df['ShipDate'] = pd.to_datetime(df['ShipDate']).dt.strftime('%d-%m-%Y')
```

```python
# Create FullName column
```

```python
df['FullName'] = df['FirstName'] + ' ' + df['LastName']
```

```python
# Change Datatype of 'PostalCode' column as string
```

```python
df['PostalCode'] = df['PostalCode'].astype(str)
```

```python
# Basic EDA (Exploratory Data Analysis)
```

```python
# Calculate Total Sales
```

```python
total_sales = df['Sales'].sum()
print(f"Total Sales: ${total_sales:,.2f}")
```

```python
# Calculate Total Quantity of Sales
```

```python
total_quantity = df['Quantity'].sum()
print(f"Total Quantity of Sales: {total_quantity}")
```

```python
# Calculate Total Profit
```

```python
total_profit = df['Profit'].sum()
print(f"Total Profit: ${total_profit:,.2f}")
```

```python
# create a column + line combo chart where Sales as bars (columns), Profit as a line and grouped by Year
```

```python
# Group by year
yearly_data = df.groupby('order_year')[['Sales', 'Profit']].sum().reset_index()

# Plot
fig, ax1 = plt.subplots(figsize=(6,5))

# Bar chart for Sales
ax1.bar(yearly_data['order_year'], yearly_data['Sales'], color='skyblue', label='Sales')
ax1.set_xlabel('Year')
ax1.set_ylabel('Sales ($)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Line chart for Profit
ax2 = ax1.twinx()
ax2.plot(yearly_data['order_year'], yearly_data['Profit'], color='red', marker='o', label='Profit')
ax2.set_ylabel('Profit ($)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Title and layout
plt.title('Year-wise Sales and Profit')
fig.tight_layout()
plt.show()
```

```python
# Create a Chart Discount vs Profit
```

```python
plt.figure(figsize=(6, 5))
sns.scatterplot(data=df, x='Discount', y='Profit', alpha=0.4)
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```

```python
# Create Sales by Year and Quarter chart
```

```python
# Group by Year and Quarter
sales_qtr = df.groupby(['order_year', 'order_quarter'])['Sales'].sum().reset_index()

# Convert quarter to string for better axis labels
sales_qtr['order_quarter'] = 'Q' + sales_qtr['order_quarter'].astype(str)

# Plot
plt.figure(figsize=(6, 5))
sns.barplot(data=sales_qtr, x='order_year', y='Sales', hue='order_quarter', palette='Set2')

plt.title('Sales by Year and Quarter')
plt.ylabel('Total Sales ($)')
plt.xlabel('Year')
plt.legend(title='Quarter')
plt.tight_layout()
plt.show()
```

```python
# Create donut chart Sales and profit by segment
```

```python
# Aggregate data
segment_data = df.groupby('Segment')[['Sales', 'Profit']].sum()

# Define labels and values
labels = segment_data.index.tolist()
sales = segment_data['Sales'].tolist()
profit = segment_data['Profit'].tolist()

# Plot setup
fig, ax = plt.subplots(1, 2, figsize=(8, 6))

# --- Donut Chart for Sales ---
ax[0].pie(sales, labels=labels, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))
ax[0].set_title('Sales by Segment')
# Add center text
ax[0].text(0, 0, 'Sales', ha='center', va='center', fontsize=12)

# --- Donut Chart for Profit ---
ax[1].pie(profit, labels=labels, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))
ax[1].set_title('Profit by Segment')
ax[1].text(0, 0, 'Profit', ha='center', va='center', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Create piechart Profit and Sales by Category
```

```python
# Group data by Category
category_data = df.groupby('Category')[['Sales', 'Profit']].sum()

# Extract data
labels = category_data.index.tolist()
sales = category_data['Sales'].tolist()
profit = category_data['Profit'].tolist()

# Plot setup
fig, ax = plt.subplots(1, 2, figsize=(8, 6))

# --- Pie Chart for Sales ---
ax[0].pie(sales, labels=labels, autopct='%1.1f%%', startangle=140)
ax[0].set_title('Sales by Category')

# --- Pie Chart for Profit ---
ax[1].pie(profit, labels=labels, autopct='%1.1f%%', startangle=140)
ax[1].set_title('Profit by Category')

plt.tight_layout()
plt.show()
```

```python
# City-wise Quantity and Profit Table descending order.
```

```python
# Group by City
city_summary = df.groupby('City')[['Quantity', 'Profit']].sum().reset_index()

# Sort by Profit descending
city_summary_sorted = city_summary.sort_values(by='Profit', ascending=False)

# Display the table
print(city_summary_sorted)
```

```python
# Create a Heatmap Profit by Region and Category
```

```python
# Create pivot table
pivot = df.pivot_table(values='Profit', index='Region', columns='Category', aggfunc='sum')

# Plot heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(pivot, annot=True, fmt='.2f', cmap='YlGnBu', linewidths=0.5)

# Titles and labels
plt.title('Profit by Region and Category')
plt.xlabel('Category')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
```

```python
# Create a Bar Chart on the Top 10 Sub-Categories by Sales
```

```python
# Group by Sub-Category and sum Sales
top_subcategories = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)

# Plot bar chart
plt.figure(figsize=(6, 5))
top_subcategories.plot(kind='bar', color='teal')

# Add labels and title
plt.title('Top 10 Sub-Categories by Sales')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create Chart on Shipping Time Distribution
```

```python
plt.figure(figsize=(6,5))
sns.histplot(df['ShippingTime'], bins=15)
plt.title('Shipping Time Distribution')
plt.xlabel('Days to Ship')
plt.show()
```

```python
# Create a chart for Monthly Sales Trend
```

```python
# Group by Month-Year and sum Sales
monthly_sales = df.groupby('order_month')['Sales'].sum().reset_index()
monthly_sales['order_month'] = monthly_sales['order_month'].astype(str)

# Plot line chart
plt.figure(figsize=(6, 5))
plt.plot(monthly_sales['order_month'], monthly_sales['Sales'], marker='o', color='steelblue')

# Formatting
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
